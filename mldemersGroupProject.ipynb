```python
artist_table = soup.find('table',class_='wikitable plainrowheaders sortable')
artist_table.prettify()
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]

for row in artist_table.findAll('tr'):
    g1 = ''
    g2 = ''
    artist=row.findAll('th')[0].find(text=True)

    cells=row.findAll('td')
    if len(cells)==6:
        for g in cells[3]:
            if isinstance(g, NavigableString) or g.name == "a":
                 g1 += g.string.strip()
                
        A.append(artist)
        B.append(cells[0].find(text=True).rstrip())
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(g1)
        F.append(cells[4].findAll('b')[0].find(text=True))
        G.append(cells[5].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Artist'])
df['Country / Market']=B
df['Period active']=C
df['Release year of first charted record']=D
df['Genre']=E
df['Total certified units (from available markets)']=F
df['Claimed sales']=G
df
```
